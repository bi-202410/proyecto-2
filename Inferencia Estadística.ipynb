# Inferencia Tumores

```python
# Carga de datos
path = '/content/drive/MyDrive/Colab Notebooks/BI/LABS/resultados_tumores.csv'
df = pd.read_csv(path)
df.head()
```

```python
df['vecesPorSemanaDeActividadFisica'].unique()
```

```python
# Mapeo de valores a reemplazar
replacement_mapping = {
    "No": 0,
    "Si": 1,
    "Sí":1,
    'No practicó deporte ni tuvo actividad física en el mes':1
}

# Reemplazar los valores en las columnas específicas
columns_to_replace = ['diagnosisDeTumoresMalignosCancer','cercaniaAIndustria', 'cercaniaABasureros','cercaniaDeGasolinearias', 'cercaniaATerminalesDeTransporte',
                      'problemasDeContaminacionDelAire', 'problemasDeDisposicionDeBasuras','haFumadoRecientemente','vecesPorSemanaDeActividadFisica']
df[columns_to_replace] = df[columns_to_replace].replace(replacement_mapping)
```

```python
df['vecesPorSemanaDeActividadFisica'].unique()
```

```python
df.head()
```

```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado df con tus datos
# df = pd.read_csv('tus_datos.csv')  # Si estás leyendo los datos desde un archivo CSV

# Definir las características y el objetivo
features = ['cercaniaAIndustria', 'cercaniaABasureros', 'cercaniaDeGasolinearias', 'cercaniaATerminalesDeTransporte']
objetivo = 'diagnosisDeTumoresMalignosCancer'

# Crear diccionarios para almacenar los resultados
promedios = {}
varianzas = {}
cantidades = {}

# Calcular los valores necesarios para cada característica
for feature in features:
    group = df.groupby(feature)[objetivo]
    promedios[feature] = group.mean()
    varianzas[feature] = group.var()
    cantidades[feature] = group.count()

# Mostrar los resultados
for feature in features:
    print(f'Característica: {feature}')
    print(f'Promedios:')
    print(promedios[feature])
    print(f'Varianzas:')
    print(varianzas[feature])
    print(f'Cantidades:')
    print(cantidades[feature])
    print()
```

```python
# Calcular para cercaniaAIndustria
group_industria = df[df['cercaniaAIndustria'] == 1][objetivo]
promedio_industria = group_industria.mean()
varianza_industria = group_industria.var()
cantidad_industria = group_industria.count()

# Calcular para cercaniaABasureros
group_basureros = df[df['cercaniaABasureros'] == 1][objetivo]
promedio_basureros = group_basureros.mean()
varianza_basureros = group_basureros.var()
cantidad_basureros = group_basureros.count()

# Calcular para cercaniaDeGasolinearias
group_gasolinearias = df[df['cercaniaDeGasolinearias'] == 1][objetivo]
promedio_gasolinearias = group_gasolinearias.mean()
varianza_gasolinearias = group_gasolinearias.var()
cantidad_gasolinearias = group_gasolinearias.count()

# Calcular para cercaniaATerminalesDeTransporte
group_terminales = df[df['cercaniaATerminalesDeTransporte'] == 1][objetivo]
promedio_terminales = group_terminales.mean()
varianza_terminales = group_terminales.var()
cantidad_terminales = group_terminales.count()

print(f'Probabilidad promedio de tener un tumor maligno para cercaniaAIndustria: {promedio_industria * 100:.4f}%')
print(f'Probabilidad promedio de tener un tumor maligno para cercaniaABasureros: {promedio_basureros * 100:.4f}%')
print(f'Probabilidad promedio de tener un tumor maligno para cercaniaDeGasolinearias: {promedio_gasolinearias * 100:.4f}%')
print(f'Probabilidad promedio de tener un tumor maligno para cercaniaATerminalesDeTransporte: {promedio_terminales * 100:.4f}%')

promedios = [promedio_industria, promedio_gasolinearias, promedio_gasolinearias, promedio_terminales]
varianzas = [varianza_industria, varianza_basureros, varianza_gasolinearias, varianza_terminales]
cantidades = [cantidad_industria, cantidad_basureros, cantidad_gasolinearias, cantidad_terminales]
```

```python
import pandas as pd
from scipy.stats import ttest_ind, f_oneway
# Crear matriz de homogeneidad
num_features = len(promedios)
homogeneidad_matriz = [[0]*(num_features+1) for _ in range(num_features+1)]

# Añadir nombres de características a la matriz
homogeneidad_matriz[0][1:] = features[:]
for i, feature in enumerate(features, start=1):
    homogeneidad_matriz[i][0] = feature

# Pruebas de homogeneidad
for i in range(1, num_features+1):
    for j in range(1, num_features+1):
        if i != j:
            # Prueba t para igualdad de medias
            t_stat, t_pvalue = ttest_ind(df[df[features[i-1]] == 1][objetivo], df[df[features[j-1]] == 1][objetivo], equal_var=False)
            # Prueba F para igualdad de varianzas
            f_stat, f_pvalue = f_oneway(df[df[features[i-1]] == 1][objetivo], df[df[features[j-1]] == 1][objetivo])

            # Si ambas pruebas no rechazan la hipótesis nula de igualdad, marcar como homogéneo
            if t_pvalue > 0.05 and f_pvalue > 0.05:
                homogeneidad_matriz[i][j] = 1

# Imprimir la matriz de homogeneidad
print("Matriz de Homogeneidad (1: Homogéneas, 0: No Homogéneas):")
for row in homogeneidad_matriz:
    print(row)
```

```python
# Calcular para problemasDeContaminacionDelAire
group_aire = df[df['problemasDeContaminacionDelAire'] == 1][objetivo]
promedio_aire = group_basureros.mean()

# Calcular para problemasDeDisposicionDeBasuras
group_disp_basura= df[df['problemasDeDisposicionDeBasuras'] == 1][objetivo]
promedio_disp_basura = group_disp_basura.mean()

print(promedio_aire*100)
print(promedio_disp_basura*100)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

featres_ambiente = ['problemasDeContaminacionDelAire', 'problemasDeDisposicionDeBasuras']
objetivo = 'diagnosisDeTumoresMalignosCancer'

# Filtrar los datos para cada característica
aire = df[df['problemasDeContaminacionDelAire'] == 1][objetivo]
basuras = df[df['problemasDeDisposicionDeBasuras'] == 1][objetivo]

# Realizar la prueba t de una cola (alternativa: media de aire > media de basuras)
t_stat, p_value = ttest_ind(aire, basuras, alternative='greater')

# Imprimir los resultados
print(f'Estadístico t: {t_stat}')
print(f'Valor p: {p_value}')
```
